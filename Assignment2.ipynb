```python
# 1. Create variables of all Python primitive data types. 
# Print their: • Value  • Data Type  • Memory Address (id()) 

a = 10          #int
print("Value: ",a,"\nDataType: ",type(a), "\nMemory Address: ", id(a))

b = 15.50          #float
print("\n\nValue: ",b,"\nDataType: ",type(b), "\nMemory Address: ", id(b))

c = True          #bool
print("\n\nValue: ",c,"\nDataType: ",type(c), "\nMemory Address: ", id(c))

d = "Hello"          #string
print("\n\nValue: ",d,"\nDataType: ",type(d), "\nMemory Address: ", id(d))

e = 15 + 4j          #complex numbers
print("\n\nValue: ",e,"\nDataType: ",type(e), "\nMemory Address: ", id(e))

f = None          #NoneType
print("\n\nValue: ",f,"\nDataType: ",type(f), "\nMemory Address: ", id(f))
```

```python
# 2. Create three variables 
# - a = 15 - b = 4 
# • Addition  • Subtraction  • Multiplication  
# • Division  • Floor Division  • Modulus  • Exponent 

a = 15
b = 4
print("a: ",a)
print("b: ",b)

res = a + b
print("Addition         : ",res)

res = a - b
print("Subtraction      : ",res)

res = a * b
print("Multiplication   : ",res)

res = a / b
print("Division         : ",res)

res = a // b
print("Floor Division   : ", res)

res = a % b 
print("Modulus          : ", res) 

res = a ** b 
print("Exponent         :  ", res)


```

```python
# 3. Create two variables: 
# - x = "100" - y = 50 
# • Convert the variables so that you can perform arithmetic operations between them.

x = "100"
y = 50 
print("Original Values: \nx: ",x,type(x),"\ny: ",y,type(y))

#Arithmetic Operations
x = int(x)
print("\nAfter Conversion:")
print("x:", x, type(x))
print("y:", y, type(y))

print("\nArithmetic Operations:")
print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)

x = float(x)
y = float(y)
print("\nAfter Conversion:")
print("x:", x, type(x))
print("y:", y, type(y))

print("\nArithmetic Operations:")
print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
```

```python
# 5. Write a program to calculate electricity bill. 
# Use variables for: • Units Consumed  • Cost per Unit  • Calculate total bill and GST (18%). 

units_consumed = 150
cost_per_unit = 8

total_bill = units_consumed * cost_per_unit
gst = total_bill * 0.18
final_bill = total_bill + gst

print("Total units consumed : ", units_consumed)
print("Cost per unit        : ", cost_per_unit)
print("Total bill           : ", total_bill)
print("GST (18%)            : ", gst)
print("Final bill           : ", final_bill)
```

```python
# 7. Write a program that converts: 
# •  int → float  •  float → string  •  string → int  •  int → bool  •  bool → int 

num = 10
print(num, type(num))
float_num = float(num)
print("int to float:", float_num, type(float_num))

decimal = 25.5
print("\n\n",decimal, type(decimal))
string_num = str(decimal)
print("float to string:", string_num, type(string_num))

text = "100"
print("\n\n", text, type(text))
int_num = int(text)
print("string to int:", int_num, type(int_num))

number = 1
print("\n\n", number, type(number))
bool_num = bool(number)
print("int to bool:", bool_num, type(bool_num))

value = True
print("\n\n", value, type(value))
int_value = int(value)
print("bool to int:", int_value, type(int_value))

```

```python
# 10. Create a program that stores product information. 
# Store: • Product Name  • Price  • Quantity  
# Calculate: • Total Cost  • GST (18%)  • Final Amount 

prod_name = "Mobile"
prod_price = 15000
prod_quantity = 10

total_cost = prod_price * prod_quantity
gst = total_cost * 0.18
final_amount = total_cost + gst

print("Product Name :", prod_name)
print("Price        :", prod_price)
print("Quantity     :", prod_quantity)
print("Total Cost   :", total_cost)
print("GST (18%)    :", gst)
print("Final Amount :", final_amount)
```
