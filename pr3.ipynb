```python
#Class Session: 23 July, 2026

#datatypes:
#1.primitive: int, float, bool, str (basic data types)

std_roll_no = 101 #int
print("Student Roll Number:", std_roll_no)

marks = 75.4 #float
print("Student Marks:", marks)

#boolean: True or False
is_passed = True #1
is_present = False #0
print("Is Student Passed:", is_passed)
print("Is Student Present:", is_present)

#string: sequence of characters
std_name = "Alice" #str
std_surname = 'Smith' #str
print("Student Name:", std_name, std_surname)

std_info = '''student name is: Alice
            and surname is: Smith''' #str
print("Student Info:", std_info)            

complex_number = 3 + 4j #complex  4j=imaginary part 3=real part
print("Complex Number:", complex_number)

#nothing
user_name = None #NoneType (explicitly null value)
print("User Name:", user_name)

```

```python
#2. Non-primitive: list, tuple, set, dict (collection data types)
#i) list: ordered, mutable, allows duplicates
std_marks = [75, 80, 90, 85, 70] #list
print("Student Marks List:", std_marks)
std_marks.append(95) #adding new element to the list 
print("Updated Student Marks List:", std_marks)

#ii) tuple: ordered, immutable, allows duplicates
std_roll_nos = (101, 102, 103, 104, 105) #tuple
print("Student Roll Numbers Tuple:", std_roll_nos)

#iii) set: unordered, mutable, does not allow duplicates
std_subjects = {"Maths", "Science", "English", "History"} #set
print("Student Subjects Set:", std_subjects)
std_subjects.add("Geography") #adding new element to the set
print("Updated Student Subjects Set:", std_subjects)

#iv) dict: unordered, mutable, key-value pairs, keys are unique
std_info_dict = {"name": "Alice", "surname": "Smith", "roll_no": 101, "marks": 75.4} #dict
print("Student Info Dictionary:", std_info_dict)
```

```python
#Type conversion: converting one data type to another data type
percentage = 75.4 #float
print("Percentage:", percentage, type(percentage), id(percentage))

temp = percentage #temp is a reference to percentage
print("Temp:", temp, type(temp), id(temp))

temp = int(percentage) #converting float to int
print("Temp after conversion to int:", temp, type(temp), id(temp), "\n")

#i) implicit type conversion: automatically converting one data type to another data type by python
#small datatype is converted to a larger data type not vice versa, if it's done vice versa then it will cause in data loss
num1 = 10           #int
num2 = 5.5          #float 
res = num1 + num2   #int is converted to float
print("Implicit:-\nResult:", res, type(res), id(res))

#ii) Explicit type conversion: manually converting one data type to another data type by using built-in functions
num3 = "10"           #string
num4 = "20"         #string

res = num3 + num4         #string concatenation
print("Explicit:-\nResult:", res, type(res), id(res))  #we wont get the sum of two numbers because they are in string format, so we need to convert them to int first

res = int(num3) + int(num4)  #converting string to int
      #10          #20
print("Result after conversion to int:", res, type(res), id(res))  #now we will get the sum of two numbers because they are in int format
```

```python
#Built-in functions: pre-defined functions in python which are available for use without importing any module

#i) type(): returns the data type of an object
num5 = 10
print("Type of num5:", type(num5))

#ii) id(): returns the unique id of an object
print("ID of num5:", id(num5)) #location of the object in memory

```
